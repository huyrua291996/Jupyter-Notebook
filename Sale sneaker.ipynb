```python
re = requests.get("https://jamja.vn/api/v3/searchdeal/?offset=0&limit=12&province=ha-noi&keyword=sneaker")
```

```python
api = json.loads(re.text)
```

```python
list_deal = api['objects']
```

```python
len(list_deal)
```

```python
for deal in list_deal:
    #deal = json.loads(deal)
    print(deal['apply_days'])
    print(deal['brand']['brand_name'])
    print(deal['condition'])
```

```python
for deal in list_deal:
    print(deal['deal_url'])    
```
